```python
# Drop the columns with morethan 10% missing value . CAUTION: This will potentailly reduce the total emission amount by Area. 

agri_food_co2_emission.drop(columns=['Crop Residues','IPPU','Manure applied to Soils','Manure Management','On-farm energy use'], inplace=True)



```

```python
# Droping the 'total-emission' variable

agri_food_co2_emission.drop(columns=['total_emission'], inplace=True)
```

The original dataset contains a variable on 'total_emission' which represents the total emission per Area. The current analysis will mainly focus on the total emission per Area. However, I chose not to use the 'total_emission' variable from the original dataset. Because the way 'total_emission' is calculated in the original variable is by simply adding the columns with non-missing values, and simply ignoring missing values. This may disproportionately affect the 'total_emission' per Area depending on missing values. I have rather chosen to exclude the variables with more than 10% missing value in the calculation of the total emission per Area. This approach would minimize the disproportionate effect of missing values in the total emission calculation.

### Data wrangling

```python
# Filter data where 'Year' is equal or greatthan 1993 and save changes inplace. This is to reduce the complication that arised from administrative chages before the year 1993 specially in USSR where some countries do not have data for the years before 1991. By focusing only the years 1993-2020 we improve data quality 

agri_food_co2_emission = agri_food_co2_emission.loc[agri_food_co2_emission['Year'] >= 1993]
```

```python
# Checking the new dataset

agri_food_co2_emission.head()
```

```python
# List all columns in the dataset to view list of columns 

column_names = agri_food_co2_emission.columns
print(column_names)
```

```python
# Derive a new variable 'Total emission' to represent the total emission from existing variables 

columns_to_sum = ['Savanna fires', 'Forest fires', 'Rice Cultivation',
       'Drained organic soils (CO2)', 'Pesticides Manufacturing',
       'Food Transport', 'Forestland', 'Net Forest conversion',
       'Food Household Consumption', 'Food Retail', 'On-farm Electricity Use',
       'Food Packaging', 'Agrifood Systems Waste Disposal', 'Food Processing',
       'Fertilizers Manufacturing', 'Manure left on Pasture',
       'Fires in organic soils', 'Fires in humid tropical forests'] # List of the relevant columns



agri_food_co2_emission['Total emission'] = agri_food_co2_emission[columns_to_sum].sum(axis=1) # Creating the new variable with axis=1 to perform operation row wise

```

```python
# Deriving a new variable 'Total population' to represent total population 

agri_food_co2_emission['Total population'] = agri_food_co2_emission['Total Population - Male']+ agri_food_co2_emission['Total Population - Female']
```

```python
# Deriving a new variable 'Per capita emission' to represent per capita emission per Area

agri_food_co2_emission['Per capita emission'] = agri_food_co2_emission['Total emission']/ agri_food_co2_emission['Total population']
```

```python
# Drop all varaibles which are not relevant for further analysis 

agri_food_co2_emission.drop(columns=[ 'Savanna fires', 'Forest fires', 'Rice Cultivation',
       'Drained organic soils (CO2)', 'Pesticides Manufacturing',
       'Food Transport', 'Forestland', 'Net Forest conversion',
       'Food Household Consumption', 'Food Retail', 'On-farm Electricity Use',
       'Food Packaging', 'Agrifood Systems Waste Disposal', 'Food Processing',
       'Fertilizers Manufacturing', 'Manure left on Pasture',
       'Fires in organic soils', 'Fires in humid tropical forests',
       'Total Population - Male',
       'Total Population - Female' ], inplace=True)
```

### Handling outliers

```python
# Remove outliers in the 'Per capita emission' column using the IQR method
# Calculate quartiles and IQR
Q1 = agri_food_co2_emission['Per capita emission'].quantile(0.25)
Q3 = agri_food_co2_emission['Per capita emission'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

agri_food_co2_emission = agri_food_co2_emission[(agri_food_co2_emission['Per capita emission'] >= lower_bound) & (agri_food_co2_emission['Per capita emission'] <= upper_bound)]

```

```python
# Check again the distribution of the new 'Per capita emission'

sns.histplot(agri_food_co2_emission['Per capita emission'], bins=100, kde = True) # Unlike total emission, huge difference interms of per capita emission is something which needs attention
```

```python
# Excluding all 'Area' which have less than 20 years of data to create a comparable dataset. 

value_counts = agri_food_co2_emission['Area'].value_counts() # Identify unique values and their counts

unique_values_to_remove = value_counts[value_counts < 20].index #  Filter out unique values with counts less than 20

agri_food_co2_emission['Area'] = agri_food_co2_emission['Area'].apply(lambda x: 'Other' if x in unique_values_to_remove else x) #  Replace  rows based on the filtered unique values

agri_food_co2_emission = agri_food_co2_emission[agri_food_co2_emission['Area'] != 'Other']  # Filter out rows with 'Other' category
```
